```python
import pandas

#Read all data
data = pandas.read_csv('data/gapminder_gdp_oceania.csv')
print(data)
```

```python
#Use column values as row headings
data = pandas.read_csv('data/gapminder_gdp_oceania.csv', index_col='country')
print(data)
```

```python
#Find out more about the dataframe
#i.e. 2 rows: Australia, New Zealand;
#12 columns, each with 2 64-bit floating point values;
#Uses 208 bytes memory
data.info()
```

```python
#Columns variable stores data about dataframe columns
print(data.columns)
```

```python
#Transpose data - Use columns as rows and rows as columns
print(data.T)
```

```python
#Get summary statistics of columns containing numerical data
#Use the argument "include='all'" to include non-numeric data
print(data.describe())
```
